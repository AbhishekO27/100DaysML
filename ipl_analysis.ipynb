# ROW FETCHING
## MASKING
yahaa humne ek fucntion banaya hai jisme hum rows fetch kr rhe using masking method , toh mask mai humne city location and winner ke Accoring value fetch kri hai

```python
def count_city_match(city,winnername):
    mask = (df['city']==city) &(df['winner']==winnername)
    return df[mask]
    
```

```python
count_city_match('Pune', 'Royal Challengers Bangalore')
```
